```python
from pyngrok import ngrok

# Get a list of active tunnels
tunnels = ngrok.get_tunnels()
# Print details of each tunnel
for tunnel in tunnels:
    print("URL:", tunnel.public_url)
    print("Protocol:", tunnel.proto)

# You can then choose the tunnel you want to disconnect based on its URL, protocol, or other details.
```
